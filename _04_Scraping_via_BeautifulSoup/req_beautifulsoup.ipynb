```python
import requests

URL = "https://www.scrapethissite.com/pages/simple/"

try:
    # Send a GET request to the website
    res = requests.get(URL, timeout=5)  # timeout → Maximum waiting time (in seconds). Here, '5' → Wait for 5 seconds.

    # Raises an HTTPError for 4xx and 5xx status codes
    res.raise_for_status()

    print("Status Code:", res.status_code)
    print("Website loaded successfully!")

except requests.exceptions.Timeout:
    print("Error: The request timed out.")
    print("Reason: The server took too long to respond.")

except requests.exceptions.ConnectionError:
    print("Error: Connection failed.")
    print("Reason: Unable to connect to the website. Check your internet connection or the website status.")

except requests.exceptions.HTTPError as e:
    print("HTTP Error:", e)
    print("Reason: The server returned an unsuccessful status code (e.g., 404 or 500).")

except requests.exceptions.TooManyRedirects:
    print("Error: Too many redirects.")
    print("Reason: The URL redirected too many times.")

except requests.exceptions.RequestException as e: #It catches any request-related error that wasn't handled by the earlier except blocks.
    print("Request Failed:", e)
    print("Reason: An unexpected error occurred while making the request.")
```

```python
headers = {   # Tells the server which browser/client is making the request.
    "User-Agent": "Mozilla/5.0"   # Tells the server which browser/client is making the request.
}

response = requests.get(URL, headers=headers)

print(response.status_code)
```

```python
import time

print("Sending request...")

response = requests.get(URL)

time.sleep(2)      # Wait for 2 seconds

print("Status Code:", response.status_code)


# Here:
    # Request is sent.
    # Program waits 2 seconds.
    # Then prints the status code.
```

```python
from bs4 import BeautifulSoup
# beautifulsoup is 'class' imported from 'module' bs4
# we will work on 'objects' of beautifulsoup to work upon

with open("../_03_Scrapping_via_requests/Scrapped_data/data1.html", "r", encoding="utf-8") as f:
    html_content = f.read()

soup = BeautifulSoup(html_content, "lxml")
```

```python
soup
```

## find()

```python
soup.find("h3")  # 1st h tag of entire page
# OR
soup.h3
```

```python
all_h3 = soup.find_all("h3")

type(all_h3)   # bs4.element.ResultSet == behaves exactly like list

for h3 in all_h3:   # loop on all_h3 info
    name = print(h3.get_text(strip=True))    # get_text() == to extract the main content from the tag  (text inside the header)
```

```python
# to find parent of h3 on the basis of tag property

for h3 in all_h3:
    print(h3.find_parent('div')['class'])  # to access a particular attribute == like here, 'class'

# Inference all the 'div' parents of 'h3' they have the same class
```

```python
# find_next -- next sibling

for h3 in all_h3:
    population = print(h3.find_next('div').select("span.country-population")[0].get_text(strip = True))  # to get span value in the class of country population 
#     OR
    print(h3.find_next('div').select("span.country-population")[0].get_text(strip = True)) # will give the same output

    print(h3.attrs)   # extracts all attributes of h3 class 
```

```python
all_countries = []
for h3 in all_h3:
    all_countries.append([name,population])
```
